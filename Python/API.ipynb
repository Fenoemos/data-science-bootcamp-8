# Public API

```python
# Run random 10 times to get 10 random jokes
# from https://official-joke-api.appspot.com/jokes/

# import function
import requests # to get request
import time # to slow requests API, use timesleep
import pandas as pd # to create dataframe
import random
```

```python
joke_lists = []

random_lists = range(1, 410)    # There are 409 list of jokes

for i in range(10):
    x = random.choice(random_lists)
    url = f"https://official-joke-api.appspot.com/jokes/{x}"
    resp = requests.get(url)
    if resp.status_code == 200:
        json_data = resp.json()
        joke_lists.append(
            (json_data["type"],
            json_data["setup"],
            json_data["punchline"],
            json_data["id"])
        )
    else:
        joke_lists.append(
            (" ",
            f"Error with {resp.status_code}",    # It error because something occur check with code status
            " ",
            " ")
        )

    time.sleep(1) #break a second (recommend 1 to 2 second)
```

```python
# create dataframe
df = pd.DataFrame(joke_lists)
# adding column name to the respective columns
df.columns =['type', 'setup', 'punchline', 'id']
# reorder column
neworder_df = df[['id', 'type', 'setup', 'punchline']]

neworder_df
```
